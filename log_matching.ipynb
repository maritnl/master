```python
import pandas as pd
import glob
```

```python
def get_stops_from_log(path):
    main_df = pd.DataFrame()
    for file in glob.glob(path):
        print(file)
        df = pd.read_csv(file)
        date = df.columns[-1]
        df = df.reset_index()
        df.columns = df.iloc[0]
        df = df.drop(0)
        df = df[['Station type', 'Time', 'Latitude', 'Longitude', 'Comment']]
        df['Date'] = date
        if main_df.empty:
            main_df = df
        else:
            main_df = pd.concat([main_df,df])

    main_df['Date'] = pd.to_datetime(main_df['Date'], dayfirst=True)
```
